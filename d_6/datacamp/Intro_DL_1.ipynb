```python
import numpy as np
```

```python
input_data = np.array([2,3])
```

```python
weights = {'node_0': np.array([1, 1]),
           'node_1': np.array([-1, 1]),
           'output': np.array([2, -1])}
```

```python
#Hidden layer.In this example we have one hidden layer
node_0_value = (input_data * weights['node_0']).sum()
node_1_value = (input_data * weights['node_1']).sum()
```

To multiply two vectors, you can use either the dot product or the cross product, depending on the desired outcome. Here's a brief explanation of both:

1. Dot Product:
The dot product (also known as the scalar product or inner product) of two vectors measures the similarity or alignment between them. It results in a scalar value.

To compute the dot product of two vectors, let's say vector A and vector B, you multiply the corresponding components of the vectors and sum them up. The formula for the dot product is:

A  B = A1 * B1 + A2 * B2 + A3 * B3 + ... + An * Bn

For example, if you have vectors A = [1, 2, 3] and B = [4, 5, 6], the dot product would be:

A  B = (1 * 4) + (2 * 5) + (3 * 6) = 4 + 10 + 18 = 32

2. Cross Product:
The cross product (also known as the vector product) of two vectors results in a new vector that is perpendicular (orthogonal) to both input vectors. It is commonly used in three-dimensional space.

The formula for the cross product of two vectors, A and B, is given as:

A x B = [A2 * B3 - A3 * B2, A3 * B1 - A1 * B3, A1 * B2 - A2 * B1]

For example, if you have vectors A = [1, 2, 3] and B = [4, 5, 6], the cross product would be:

A x B = [(2 * 6 - 3 * 5), (3 * 4 - 1 * 6), (1 * 5 - 2 * 4)] = [12 - 15, 12 - 6, 5 - 8] = [-3, 6, -3]

Please note that the cross product is only applicable to three-dimensional vectors.

Remember to consider the dimensions and properties of the vectors to determine which type of multiplication is appropriate for your specific use case.

```python
hidden_layer_values = np.array([node_0_value, node_1_value])
print(hidden_layer_values)
```

```python
output = (hidden_layer_values * weights['output']).sum()
print(output)
```
