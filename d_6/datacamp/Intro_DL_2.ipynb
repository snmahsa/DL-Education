```python
import numpy as np
```

```python
input_data = np.array([-1, 2])
```

```python
# The weights that enter each node
weights = {'1hiden_node_0':np.array([3, 3]),
          '1hiden_node_1': np.array([1, 5]),
          'output': np.array([2, -1])}
```

```python
value_1hidden_node_0 = (input_data * weights['1hidden_node_0']).sum()
value_1hidden_node_0 = np.tanh(value_1hidden_node_0)
value_1hidden_node_1 = (input_data * weights['1hidden_node_1']).sum()
value_1hidden_node_1 = np.tanh(value_1hidden_node_1)
```

```python
hidden_layer_values = np.array([value_1hidden_node_0 , value_1hidden_node_1])
```

```python
output = (hidden_layer_values * weights['output']).sum()
print(output)
```
